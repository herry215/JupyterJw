```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import statsmodels.api as sm
```

```python
redwine = pd.read_csv('winequality-red.csv', sep=";", header=0)
redwine["type"] = "red"
redwine.head()
redwine.shape
```

```python
whitewine = pd.read_csv('winequality-white.csv', sep=";", header=0)
whitewine["type"] = "white"
whitewine.head()
```

```python
wine = redwine.append(whitewine)
wine.shape
```

```python
wine.columns = wine.columns.str.replace(" ","_")
wine.head()
```

```python
wine.describe() #요약 통계량 --카운트 평균 표준편차 min,max 사분의수 iqr 25=Q1 50=Q2 75=Q3
```

```python
wine.quality.describe()
```

```python
sorted(wine.quality.unique())
```

```python
wine.quality.value_counts()
```

```python
wine.groupby("type")["quality"].describe() #dataFrame Groupby 요약 통계량결과 출력
```

```python
wine.groupby("type")["quality"].quantile([0, 0.25, 0.5, 0.75, 1]).unstack("type") #qunatile=list를 구분해  Q1~Q5 5개의 변수정보가 담김
```

```python
red_q = wine.loc[wine["type"] == "red", "quality"]
white_q = wine.loc[wine["type"] == "white", "quality"]

sns.set_style("darkgrid")
sns.distplot(red_q, norm_hist=True, kde=False, color="red", label="Red Wine")
sns.distplot(white_q, norm_hist=True, kde=False, color="blue", label="White Wine")
plt.title("Distribution of Quality of Wine Type")
plt.xlabel("Quality Score")
plt.ylabel("Density")
plt.legend();
```

### 귀무가설 정의 -> 표본 추출 -> 통계적 유의성이 있는지 검정 기법을 이용하여 -> 결과에 따라 통계적으로 유의 한지 아닌지 (기각/채택)인지 결정한다.  (모집단위가 두개[red, white] 품질의 평균 차이(있다/없다) Anova 3개 이상 T-test 2개 이상)
### 귀무가설 : "와인 종류별 품질의 차이는 없다." 평균의 차이가 없다.
### p-value 0.05보다 작으면 유의성 없고 크면 유의성이 있다.

```python
wine.groupby("type")["quality"].aggregate(["std", "mean"]) #그룹별 집합 함수들에 대한 결과 확인. aggregate
```

```python
#t_stat, p_value, df = sm.stats.ttest_ind(red_q, white_q) 
#"t-stat: {:.3f}, p-value: {:.4f}".format(t_stat, p_value) # 0.0000이므로써 평균의 차이가 있음을 알수있다.
red_q = wine.loc[wine["type"] == "red", "quality"]
white_q = wine.loc[wine["type"] == "white", "quality"]
```

```python
sm.stats.ttest_ind(red_q, white_q)[1] <= 0.05
```

```python
red_sample = wine.loc[wine["type"] == "red", :]
white_sample = wine.loc[wine["type"] == "white", :]
```

```python
red_idx = np.random.choice(red_sample.index, replace=True, size=200)
white_idx = np.random.choice(white_sample.index, replace=True, size=200)
```

```python
wine_sample = red_sample.loc[red_idx, ].append(white_sample.loc[white_idx, ])
wine_sample.head()
```

```python
wine_sample = wine_sample.reset_index(drop=True)
sns.set_style("dark")
sns.pairplot(wine_sample, # pairplot 스캐터플롯과 히스토그램을 그려준다.
             vars=["quality", "alcohol", "residual_sugar"], 
             kind="reg", # regression을 지정하면 회귀선과 관련된 부분들을 표시해서 전체적인 경향성을 파악 가능
             plot_kws={"ci": False, "x_jitter": 0.25, "y_jitter": 0.25}, 
             diag_kind="hist",
             diag_kws={"bins": 10, "alpha": 1},
             hue="type", 
             palette=dict(red="red", white="blue", markers=["o", "s"]));
```
